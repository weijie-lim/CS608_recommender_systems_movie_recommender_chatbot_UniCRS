Extract Entity to ID

```python
import json
import os
import re


##### Try to extract from dbpedia raw instead of entity
def get_entities_and_ids():
  # Opening JSON file
  with open('/content/drive/MyDrive/UniCRS/UniCRS-main/src/data/redial/entity2id.json') as json_file:
      data = json.load(json_file)

  entity_id = {}
  id_entity = {}
  for k, v in data.items():
    k = k.split("/")[-1]
    k = k.replace(">", "")
    if "(" and ")" in k:
      k=re.sub("[\(\[].*?[\)\]]", "", k)
    k = k.replace("_", " ")
    k = k.strip()
    k = k.lower()

    entity_id[k] = v
    id_entity[v] = k

  print("SAMPLE entity_id", list(entity_id.keys())[0:10])
  for i in list(entity_id.keys())[0:10]:
    print(i, i.split(" "))
  print("SAMPLE id_entity", list(id_entity.keys())[0:10])
  return entity_id, id_entity

##### Try to extract from dbpedia raw instead of entity
def get_entities_and_ids_from_dbpedia():
  # Opening JSON file
  final = {}
  with open('/content/drive/MyDrive/UniCRS/UniCRS-main/src/data/redial/train_data_dbpedia_raw.jsonl') as json_file:
    json_list = list(json_file)

  for line in json_list:
    data = json.loads(line)
    data = data['movieMentions']
    final.update(data)

  with open('/content/drive/MyDrive/UniCRS/UniCRS-main/src/data/redial/test_data_dbpedia_raw.jsonl') as json_file:
    json_list = list(json_file)

  for line in json_list:
    data = json.loads(line)
    data = data['movieMentions']
    final.update(data)

  with open('/content/drive/MyDrive/UniCRS/UniCRS-main/src/data/redial/valid_data_dbpedia_raw.jsonl') as json_file:
    json_list = list(json_file)

  for line in json_list:
    data = json.loads(line)
    data = data['movieMentions']
    final.update(data)

  dict2 = {}
  for k, v in final.items():
      dict2[v] = dict2.get(v, []) + [k]

  return dict2, final
```

```python
import random
entity_id, id_entity = get_entities_and_ids()
all_entities = list(entity_id.keys())
apologise_error_404 = [
    "Sorry, I don't seem to know the person or movie you are talking about!",
    "Ah, my circuits seem to have been fried. I can't find the movie / person you are referring to!",
    "Sorry, error 404, that movie / person has not been found.",
    "I apologise, but can you please check your spelling. I can't comprehend what you are saying!",
    "Please try again, but with better spelling this time. Thanks :)"
]

def input_to_jsonl(input_str):
  single_user_path = '/content/drive/MyDrive/UniCRS/UniCRS-main/src/data/redial/sample_input_data_processed.jsonl'
  isExisting = os.path.exists(single_user_path)
  current_str = ''
  if isExisting:
    with open(single_user_path, 'r') as json_str:
      current_str = json.load(json_str)
      print(f"current_str: {current_str}")
      #add in new input str from user
      current_str['context'].append(input_str)
  else:
    current_str = {"context": [input_str], "resp": "", "rec": [], "entity": []}


  doesMovieExist = True
  #check string for special regex: regex indicates movies $Angelina Jolie$ $Transformers$
  result = re.findall('\$.*?\$', input_str)
  if result != []:
    result = [i.replace('$', '').lower() for i in result]
    movie_ids = []


    print("************************************************************")
    print("MOVIE IS", result)
    print("************************************************************")
    for r in result:
      if r not in all_entities:
        doesMovieExist = False
      for k, v in entity_id.items():
        if k == r:
          print("************************************************************")
          print("MOVIE ID IS", v)
          print("************************************************************")
          current_str['rec'].append(v)
  # current_str['entity'] = [int(i) for i in current_str['entity']]

  with open(single_user_path, 'w') as outfile:
    jout = json.dumps(current_str)
    outfile.write(jout)
  outfile.close()
  return doesMovieExist

def from_pred_output_resp_to_input():
  single_user_path = '/content/drive/MyDrive/UniCRS/UniCRS-main/src/data/redial_gen/sample_input_data_processed.jsonl'
  pred_reply = "/content/drive/MyDrive/UniCRS/UniCRS-main/src/save/redial/best_sample_input.jsonl"
  with open(pred_reply, 'r') as json_str:
    pred_str = json.load(json_str)
    print(f"pred_str: {pred_str}")

  curr_str = {}
  with open(single_user_path, 'r') as outfile:
    curr_str = json.load(outfile)
    print(f"curr_str: {curr_str}")
  outfile.close()

  curr_str['resp'] = pred_str['pred']

  with open(single_user_path, 'w') as outfile:
    jout = json.dumps(curr_str)
    outfile.write(jout)
  outfile.close()


def run_inference():
  # infer
  !python3 infer_conv_final.py --dataset redial --split sample_input --tokenizer microsoft/DialoGPT-small --model microsoft/DialoGPT-small --text_tokenizer roberta-base --text_encoder roberta-base --n_prefix_conv 20 --prompt_encoder /content/drive/MyDrive/UniCRS/UniCRS-main/src/conversation_prompt/best/ --per_device_eval_batch_size 64 --context_max_length 200 --resp_max_length 183 --prompt_max_length 200 --entity_max_length 32
  %pwd
  %cp -r data/redial/. data/redial_gen/
  %cd data
  %cd redial_gen
  !python merge.py --gen_file_prefix best
  from_pred_output_resp_to_input()
  %pwd
  %cd ..
  %cd ..
  !python3 infer_rec_final.py \
    --dataset redial_gen \
    --tokenizer microsoft/DialoGPT-small \
    --model microsoft/DialoGPT-small \
    --text_tokenizer roberta-base \
    --text_encoder roberta-base \
    --n_prefix_rec 10 \
    --prompt_encoder /content/drive/MyDrive/UniCRS/UniCRS-main/src/recommendation_task/best \
    --num_train_epochs 5 \
    --per_device_train_batch_size 64 \
    --per_device_eval_batch_size 64 \
    --gradient_accumulation_steps 1 \
    --num_warmup_steps 530 \
    --context_max_length 200 \
    --prompt_max_length 200 \
    --entity_max_length 32 \
    --learning_rate 1e-4 \
    --output_dir /content/drive/MyDrive/UniCRS/UniCRS-main/src/test_recommendation_task/ \
    # --use_wandb \  # if you do not want to use wandb, comment it and the lines below
    # --project crs-prompt-rec \  # wandb project name
    # --name xxx  # wandb experiment name

def read_from_jsonl_to_user():
  single_user_path = '/content/drive/MyDrive/UniCRS/UniCRS-main/src/data/redial_gen/sample_input_data_processed.jsonl'
  isExisting = os.path.exists(single_user_path)
  if isExisting:
    with open(single_user_path, 'r') as json_str:
      current_str = json.load(json_str)
      print(f"current_str: {current_str}")

  movie_count = 0
  current_str['resp'] = current_str['resp'].replace('System: ', '')
  if "<movie>" in current_str['resp']:
    split_str = current_str['resp'].split("<movie>")
    final_str = []

    for idx, j in enumerate(split_str):
      if idx != len(split_str)-1:
        final_str.append(j)
        if current_str['entity'] == []:
          final_str.append(random.choice(all_entities))
        else:
          final_str.append(id_entity[current_str['entity'][-movie_count-1]])
        movie_count +=1
      else:
        final_str.append(j)
    return ' '.join(final_str)
  else:
    return current_str['resp']

def full_procedure(input_str):
  doesEntityExist = input_to_jsonl(input_str)
  print("***************")
  print("Does entity exist:", doesEntityExist)
  print("***************")
  if doesEntityExist:
    run_inference()
    return read_from_jsonl_to_user()
  else:
    return random.choice(apologise_error_404)
```

```python
from env import API_TOKEN
import telebot
bot = telebot.TeleBot(API_TOKEN)

welcome_message = """Hi there! I am Dr. Popcorn.
I am a sociable bot, even though sometimes I don't make sense!\n
For movie recommendations, please enclose the movies in $...$
For instance:
I love to watch movies like $American Pie$. Can you recommend me movies like that?\n
Hope you find me interesting enough, so that my creators can get an A+ for their Project :)\n
Cheers,
Dr Popcorn
"""

lemme_check = """
****************************
  Thanks for your message!
****************************

         Please Hold...
    Sorry, I'm a little slow
          sometimes...

****************************

Cheers,
Dr Popcorn
"""

restricted_access = ["weijieeee", "Jkcol", "studying", "EvilCinderella"]
# restricted_access = ["weijieeee"]

# Handle '/start' and '/help'
@bot.message_handler(commands=['start'])
def send_welcome(message):
  print(message.from_user.username, "just logged on!")
  bot.reply_to(message, welcome_message)
  bot.send_message(message.chat.id,lemme_check)

# Handle '/start' and '/help'
@bot.message_handler(commands=['clear_cache'])
def clear_garbage(message):
  print(message.from_user.username, "cleared cache")
  msg = clear_cached()
  bot.reply_to(message, msg)

# Handle all other messages with content_type 'text' (content_types defaults to ['text'])
@bot.message_handler(func=lambda message: True)
def echo_message(message):
  print(message.from_user.username, "said/asked:", message.text)
  if message.from_user.username in restricted_access:
    replies = full_procedure(message.text)
    bot.reply_to(message, replies)
  else:
    bot.reply_to(message, "Sorry, you got no access!")



bot.infinity_polling()
```
